```python
import pandas as pd
import numpy as np
#creating series from array
arr=np.array(['p','a','n','d','a','s'])
a=pd.Series(arr)
print("Series from array")
print(a)
```

```python
#series from dictionaryimport pandas as pd
import numpy as np
#creating series from array
arr=np.array(['p','a','n','d','a','s'])
a=pd.Series(arr)
print("Series from array")
print(a)
arr={'x':0.,'y':1.,'z':2.}
b=pd.Series(arr)
print("\n\nseries from dictionary:\n")
print(b)
```

```python
#series using scalar
x=pd.Series(4,index=[0,1,2,3])
print("\nserve using scalar\n")
print(x)
```

```python
#series through index
x=pd.Series([1,2,3],index=['a','b','c'])
print("\nseries through index")
print(x)
```

```python
a=pd.Series(data=[1,2,3,4])
print("\nseries:\n",a)
print("\nIndex:\n",a.index)
print("\nValue:\n",a.values)
print("\nShape:\n",a.shape)
print("nDimension:\n",a.ndim)
print("\nSize\n",a.size)
```
